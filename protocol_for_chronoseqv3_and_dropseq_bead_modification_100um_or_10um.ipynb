### Calculations and Notes:

+ The scale of synthesis for each SEQ or Batch of Dropseq Beads is 10&mu;mol.
+ We will divide these beads into 1/4 parts for each single Time-tag. This is assuming we want 16 Time-Tags from SEQ1-SE4 ChronoSeq V3 or 4 from 10&mu;mol Drop-Seq beads.
+ So about 2.5&mu;mol per Time-tag. 10&mu;mol of Beads comes to about 12-13mls at 450beads/&mu;l. This comes to about 3 mls of beads at 450beads/&mu;l for each 1/4 Split for a Time-Tag.
+ We have previously done an [experiment](protocol_for_exo1_predigestion_for_chronoseqV2_and_dropseq_beads.ipynb) with 0.4ml of beads at 450beads/&mu;l. This should be equivalent to about 0.25&mu;mol.
+ This previous experiment we had no problems when we did the Exo 1 Digestion.
+ We have several things to keep in mind as constraints for our future experiments now.
+ If we do the whole 2.5&mu;mols we should expect to have about 1.25&mu;mol of complete primers and 2.5&mu;mols including incomplete primers. Incomplete primers will be relevant for Exonuclease 1 Digestion.
+ So if we assume we will be synthesizing 100bases for each complete sequence we need 125&mu;mols of total dNTPs. But since the reaction mixture is equimolar and we need about half our bases to be dTs we need about half or 50&mu;mols of dNTPs.**Thats a lot. Too much infact.**
+ 6&mu;mols for the 0.4ml fraction at least. But we used a much smaller number of primers in our previous experiment and this did not create any problems (Exo1 + Alkaline Denaturation) compared to unmodified Drop-Seq. So from now onwards we will use the primer molar amount for our calculations as a baseline.

#### Calculations for 0.25&mu;mol/0.4ml Test Batch:

+ So for our test reaction with 0.25&mu;mol.
+ 20&mu;l of 1mM primers is 10nmol of primers.
+ So for 100 bases we need about 1000 nmol of dNTPs or 500nmol for dTs in particular. T(30) + 70/4. so 0.5&mu;mol total of eqimolar primers. 100&mu;l of [dNTP mix](https://www.neb.com/en-us/products/n0447-deoxynucleotide-dntp-solution-mix#Product%20Information) should be enough for that. We can titrate this down if we see equally good results for lower amounts of primers. Just multiply the primer amount by 100 and that should give us the amount of dNTPs needed. 
+ We also need a Polymerase that is fast enough to get this done and do it at a temperature lower than the melting temperature of our ChronoSeq V3 beads.
+ phi29 DNA polymerase produces blunt ends but is too slow. [Only 0.5pmol of dNTPs at 30&deg;C in 10 minutes.](https://www.neb.com/en-us/products/m0269-phi29-dna-polymerase#Product%20Information_Properties%20&%20Usage)
+ We might use Bsu DNA polymerase. The optimum reaction temperature for this polymerase is lower than our sticker sequence melting temperatures at 37&deg;C. Moreover the enzyme keeps 50% of its activity at 25&deg;C which is still pretty fast. However, this enzyme produces [3'A single base overhangs](https://www.neb.com/en-us/products/m0330-bsu-dna-polymerase-large-fragment#Product%20Information_Properties%20&%20Usage) which can cause problems later.
+ [DNA Polymerase I](https://www.neb.com/en-us/products/m0209-dna-polymerase-i-e-coli#Product%20Information) produces blunt ends. Moreover, the optimum reaction temperature for this polymerase is lower than our sticker sequence melting temperatures at 37&deg;C. However, we need to be careful in deciding the reaction conditions to avoid getting recessed ends.
+ According to the [NEB Website](https://www.neb.com/en-us/products/m0209-dna-polymerase-i-e-coli#Product%20Information_Properties%20&%20Usage) at 37&deg;C we get 10nmol of dNTPs incoporated by one unit of DNA Polymerase I in 30 minutes. There are 10units/&mu;l and we will use 5&mu;l of polymerase. So that gives us 500nmol of dNTPs per 30 minutes. This reaction should be complete in 1 hour at 37&deg;C. We might need to increase this time for larger amount of primer or beads in our reaction mixture in the future. 
+ Lets run this reaction at 34&deg;C assuming 75% activity levels. We will run this reaction for 1.5 hours. This should be enough time to complete the reaction and [avoid recessed ends](https://www.neb.com/en-us/products/m0210-dna-polymerase-i-large-klenow-fragment#Product%20Information_Product%20Notes). We assume the Polymerase will have 25% Activity levels similar to the DNA Polymerase I, Large Klenow Fragment since [direct information for this Polymerase is not available](https://www.neb.com/en-us/products/m0330-bsu-dna-polymerase-large-fragment#Product%20Information_Product%20Notes). 100% at 37&deg;C and 25% at 25&deg;C gives 75% Activity at 34&deg;C using a linear approximation. 
+ We will then run an [Exonuclease 1](https://www.neb.com/en-us/products/m0293-exonuclease-i-e-coli#Product%20Information) to get rid of the primers that were not extended.
+ According to the [NEB website](https://www.neb.com/en-us/products/m0293-exonuclease-i-e-coli#Product%20Information_Properties%20&%20Usage) 1 unit of Exonuclease 1 removed 10nmol at 37&deg;C in 30 minutes. There are 20 units/&mu;l and we will use 50&mu;l for our reaction. So we can remove 20&mu;mol of dNTPs in 1 hour. For 2.5&mu;mol reaction we need to remove about 56-82 bases depending on if its Dropseq or Chronoseq V3. 210&mu;mol for Dropseq considering only a small fraction of sequences will actually have the primer. We will have to run this reaciton for 11 hours in that case or we could use more enzyme. But for the 0.25&mu;mol reaction we need to remove at most 21&mu;mol. So in 1.5 hours the reaction should be complete at 37&deg;C. So our previous attempt should have completely gotten rid of these single stranded primers.
+ After this step we need to get rid of the primer through Alkaline denaturation similar to before with the Exonuclease 1 and Alkaline denaturation protocol. We also need to maintain the salt concentration to make sure the primers don't get removed before the Alkaline Denaturation step. We want the Exonuclease 1 to act only on ssDNA and not dsDNA which is why this step is important.

### Protocol for Converting dsDNA on the Beads to ssDNA [using Alkaline Denaturation](http://dx.doi.org/10.5620/eht.2014.29.e2014007)
+ Sodium Hydroxide is a very dangerous Chemical it can melt your Skin and turn your body into Soap. You need to be extra careful.
+ Wear Lab Coat, Safety Glasses, [Apron](https://www.mcmaster.com/50085T13/), Face-Sheild, Double Nitrile Gloves and work in the Fume Hood. 
+ Make sure you don't have exposed skin by covering your gloves with the apron.
+ Replace the work area mats with [Fresh new ones](https://us.vwr.com/store/product/11988983/aqui-padtm-benchtop-absorbent-mat-therapak).
+ Make sure there is enough space to work inside the Fume hood. 
+ Bring the [10M NaOH Bottle](https://www.fishersci.com/shop/products/sodium-hydroxide-10n-aq-soln-thermo-scientific/AAJ63736AE) stored under your bench in a secondary container to the Fume Hood. You should always store the bottle inside a secondary container inside another larger container.
+ You will need one Hazardous Waste Bin for the Tips. Bring The Black Waste Bin into The Fume Hood. Check the Waste Tag to make sure this is the correct Bin.
> &#128161; **Tip:** Make sure you have enough volume of Denaturation Solution and 1M Tris-HCl pH 7.5 for your experiments.
+ You will also need a Buffer to Dispose off the 1M NaOH washed through.
+ For this purpose I recommend having between 30-40ml of [1M Tris-HCl pH 7.5](https://www.fishersci.com/shop/products/molecular-biology-reagents/MT46030CM) in a 50ml Tube Ready. Label this Tube: **Neutralization**.
+ Set aside enough of 1M Tris-HCl pH 7.5 in a different tube for washing the Beads. Label this Tube: **Washing**.
+ You will also need to prepare a Spray Bottle full of 1M Tris-HCl pH 7.5 Buffer. This will act as our emergency spray in case of a spill. Label This bottle correctly and Keep it inside the Fume Hood. Very important to have this with you.
> &sext; **Important**: Discard the tips used for Handling NaOH into the Black Hazardous Waste Bin.
> <br>**In case of a Spill**: Spray all affected surfaces with the 1M Tris-HCl pH 7.5.
+ We Will now make Fresh Denaturation Solution. To Make 1ml of Denaturation Solution:
    + 900&mu;l of Distilled Water
    + 100&mu;l of 10N NaOH.
    + Pipette up and down to mix well.
    + Discard the NaOH contaminated tip into the Black Bin.
+ Once you have made enough Denaturation Solution, put the 10M NaOH back in its storage location inside its secondary container. Make sure the 10M NaOH bottle is closed tightly.
+ Change your gloves.
+ Now wash the beads thrice with 1ml Denaturation Solution:
