# Protocol for Making ChronoSeq V4 Beads by Modifying Dropseq beads for Bulk Time Series.
This protocol is for making 400&mu;l of Beads for each Time-Tag. Ideal for Testing in Bulk Time Series.

|<h3>Bead Modification Workflow</h3>|
|-|
|<img src="img/Copy%20of%20ChronoSeq%20Bead%20Synthesis.png">|

Sequences are added by using an oligo that binds to the existing PolyT region on Drop-seq beads. The 3’ end of the Drop-seq beads are then extended to the reverse compliment of the oligo using E. Coli DNA Polymerase I. Single-stranded DNA are removed using Exonuclease 1. This leaves only double-stranded DNA on the beads that cannot capture RNA. Therefore, Alkaline denaturation is used to make the double stranded DNA single-stranded. These beads are then resuspended in lysis buffer and can be used directly with our device.

### Protocol for Bead Modification
#### Preparation of Beads
+ [Use the instructions in this section to](protocol_library_preparation_for_dropseq_chronoseq_beads.ipynb#How-to-Dissolve-Primers/Oligos-in-IDTE) make 10&mu;M stocks for the Sticker Oligos ordered from IDT, using the IDTE buffer. 
+ Follow the [ChronoSeq beads preparation protocol](protocol_for_preparing_chronoseq_dropseq_beads_stored_in_ethanol.ipynb) and suspend the Dropseq Beads from the Alcohol stock at -20&deg;C to [1X Lysis buffer](protocol_for_preparing_3.33X_lysis_buffer_stock.ipynb) at 4&deg;C.
    + You will need about 400&mu;l of beads @450beads/&mu;l per Time-Tag for this protocol.
+ Spin Down Dropseq beads with 450beads/&mu;l suspended in 1X Lysis buffer.
+ Turn on the UV and airflow for the Cell Culture Hood. Leave for 15 minutes.
+ Inside the Cell Culture Hood: Shake Each Tube to make sure the beads are <b>evenly suspended</b> . For each Tube using a <b>Sterile/Particle Free</b> [Rainin P1000 pipette tip](https://www.shoprainin.com/Pipette-Tips/Universal-Pipette-Tips/Hinged-Cover-Racks/Pipette-Tips-RT-UNV-1000%C2%B5L-LS-768A-8/p/30389169) transfer 400&mu;l of beads to a <b>Sterile</b> 1.5ml Eppendorf Tube.
+ Put the main 50ml Tubes with Dropseq beads back in their storage location at 4&deg;C once you are done.
+ Turn off the Cell Culture Hood and move to your bench. Take the eppies with 400&mu;l beads at 450beads/&mu;l with you.
> This is a **STOPPING STEP**. You can store the Beads in 1X Lysis Buffer at 4&deg;C and continue later.

#### DNA Polymerase I Isothermal Extension
+ Get the eppies with 400&mu;l of Dropseq beads suspended in 1X Lysis buffer.
+ Set an incubator to 34&deg;C
+ Get the dNTP, 10&mu;M Sticker Oligos and [NEB Buffer 2](https://www.neb.com/en-us/products/b7002-nebuffer-2#Product%20Information) from -20&deg;C and leave on your bench to melt. Put on ice once melted.
> &#128161; **Tip:** Make sure you have enough volume of dNTPs, Oligos and Buffer for your experiments.
+ Get the DNA Polymerase I from -20&deg;C and put it on ice.
+ Make 1.25X NEB Buffer 2 and put on ice.
+ Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
+ Remove 300&mu;l of Lysis buffer without disturbing the beads.
+ Now wash the beads 3 Times with 1ml 1.25X NEB Buffer 2:
    + Add 1ml 1.25X NEB Buffer 2. Vortex the beads at full speed for 2 seconds to resuspend the beads if necessary.
    + Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
    + Carefully remove 1ml of Liquid without disturbing the beads.
    + Add 1ml 1.25X NEB Buffer 2. Vortex the beads at full speed for 2 seconds to resuspend the beads if necessary.
    + Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
    + Carefully remove 1ml of Liquid without disturbing the beads.
    + Add 1ml 1.25X NEB Buffer 2. Vortex the beads at full speed for 2 seconds to resuspend the beads if necessary.
    + Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
+ Without disturbing the beads remove 400&mu;l of Buffer.
+ Estimate the Volume using the P1000.
+ Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
+ Remove liquid without distrubing the beads to make the final Volume 320&mu;l.
+ Now add:
    + 20&mu;l of 10&mu;M Sticker Oligo for ChronoSeq-V4. Pipette up and down several times to mix well.
    + 20&mu;l dNTP mix from NEB
    + 39&mu;l [Distilled Water](https://www.fishersci.com/shop/products/invitrogen-ultrapure-dnase-rnase-free-distilled-water-2/10977023)
    + 1&mu;l DNA Polymerase I . <em>Use the P2.5 , Pipette up and down several times to mix well.</em>
+ Vortex the Beads at full speed for 2 seconds to mix everything well.
+ Now Incubate at 34&deg;C with Rotation for 1.5 hours.
+ Set the incubator to 37&deg;C after you are done using it.
+ Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
+ Without disturbing the beads remove 200&mu;l of Buffer.
+ Wash the beads once with STOP Buffer and then twice with WASH Buffer:
    + Add 1 ml of STOP Buffer to the eppie.
    + Flip the tube upside down a few times and then Vortex the beads at full speed for 2 seconds. 
    + Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
    + Carefully remove 1ml of Liquid without disturbing the beads.
    + Add 1ml of WASH Buffer. Vortex the beads at full speed for 2 seconds to resuspend the beads if necessary.
    + Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
    + Carefully remove 1ml of Liquid without disturbing the beads.
    + Add 1ml of WASH Buffer. Vortex the beads at full speed for 2 seconds to resuspend the beads if necessary.

> This is a **STOPPING STEP**. You can store the Beads in WASH Buffer at 4&deg;C and continue later.

#### Exonuclease 1 Digestion
+ Make sure the incubator is Set to 37&deg;C.
+ Get the [Exonuclease 1 Buffer](https://www.neb.com/en-us/products/b0293-exonuclease-i-reaction-buffer#Product%20Information) -20&deg;C and leave on your bench to melt. Put on ice once melted.
> &#128161; **Tip:** Make sure you have enough volume of Exonuclease 1 Buffer for your experiment.
+ Get the Exonuclease 1 from -20&deg;C and put it on ice.
+ Make 1.25X Exonuclease 1 Buffer and put it on ice.
+ Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
+ Wash twice with 1ml of 1.25X Exonuclease 1 Buffer:
    + Carefully remove 1ml of Liquid without disturbing the beads.
    + Add 1 ml of 1.25X Exo 1 buffer to the eppie. Vortex the beads at full speed for 2 seconds to resuspend the beads if necessary.
    + Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
    + Carefully remove 1ml of Liquid without disturbing the beads.
    + Add 1 ml of 1.25X Exo 1 buffer to the eppie. Vortex the beads at full speed for 2 seconds to resuspend the beads if necessary.
    + Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute. 
+ Remove 400&mu;l of Liquid from the eppie without disturbing the beads.
+ Without touching the beads, slightly pipette up and down using the P1000 to resuspend the beads in solution.
+ Now estimate the Volume in the eppie by using the P1000.
+ Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
+ Using the estimated volume in the eppie, make the final volume of the eppie 320&mu;l without disturbing the beads.
+ Then add the following to the eppie :
    + 30&mu;l Distilled Water
    + 50&mu;l Exonuclease 1 . Pipette up and down many times to mix well.
+ Vortex the Beads to Mix everything Well.
+ Now Incubate at 37&deg;C with Rotation for 2 hours.
+ Set the incubator to 42&deg;C after you are done using it.
+ Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
+ Without disturbing the beads remove 200&mu;l of Buffer.
+ Wash the beads once with TE-SDS and then twice with TE-TW:
    + Add 1 ml of TE-SDS to the eppie.
    + Flip the tube upside down a few times and then Vortex the beads at full speed for 2 seconds. 
    + Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
    + Carefully remove 1ml of Liquid without disturbing the beads.
    + Add 1ml of TE-TW. Vortex the beads at full speed for 2 seconds to resuspend the beads if necessary.
    + Spin down the beads using a tabletop centrifuge @ 2500xg for 1 minute.
    + Carefully remove 1ml of Liquid without disturbing the beads.
    + Add 1ml of TE-TW. Vortex the beads at full speed for 2 seconds to resuspend the beads if necessary.
    
> This is a **STOPPING STEP**. You can store the Beads in TE-TW at 4&deg;C and continue later.

> This is a **STOPPING STEP**. You can store the Beads in 1X Lysis Buffer at 4&deg;C and continue later.
+ Follow the [Bulk RNA Seq protocol](protocol_for_bulk_rna_seq_library_prep.ipynb) to verify your beads are working or not.
+ Follow the [Bulk RNA Seq section for the ChronoSeq device](protocol_and_software_for_running_chronoseq_device.ipynb#bulk_cell) if you plan to do a time-series assay.
