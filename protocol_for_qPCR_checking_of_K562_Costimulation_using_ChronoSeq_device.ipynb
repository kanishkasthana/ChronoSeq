## Protocol for qPCR Checking of K562 Costimulation from THP-1 Cells

### Reagents to purchase
+ [SingleShot Cell Lysis Kit, 100 x 50 µl rxns #1725080](https://www.bio-rad.com/en-us/sku/1725080-singleshot-cell-lysis-kit-100-x-50-ul-rxns?ID=1725080)
+ [iTaq™ Universal SYBR® Green One-Step Kit, 100 x 20 µl rxns, 1 ml #1725150](https://www.bio-rad.com/en-us/sku/1725150-itaq-universal-sybr-green-one-step-kit-100-x-20-ul-rxns-1-ml?ID=1725150)
+ [PrimePCR™ SYBR® Green Assay: NFKBIA, Human](https://commerce.bio-rad.com/en-us/prime-pcr-assays/assay/qhsaced0042829-primepcr-sybr-green-assay-nfkbia-human)
+ [PrimePCR™ SYBR® Green Assay: ACTB, Human](https://commerce.bio-rad.com/en-us/prime-pcr-assays/assay/qhsaced0036269-primepcr-sybr-green-assay-actb-human)

### Instruments used: 
+ Bio-Rad CFX Opus 96 

### Cell Culture and Cell Preparation
+ We will need to culture THP-1 cells and K562 Cells for this experiment. 
+ Make sure you have enough K562 Cells on the day of the experiment to make at least 7ml of K562 at 430cells/&mu;l
+ We will follow the [Bowdish lab protocol for culturing THP-1 Cells](http://www.bowdish.ca/lab/wp-content/uploads/2011/07/THP-1-propagation-culture.pdf)
+ Make complete growth media with 10%FBS, 1%PEN-STREP, and 0.05mM (1X)[2-Mercaptoethanol](https://www.fishersci.com/shop/products/gibco-2-mercaptoethanol-55mm/21985023) with [ATCC Formulation of RPMI 1640](https://www.fishersci.com/shop/products/gibco-rpmi-1640-medium-atcc-modification-hepes-liquid/A1049101).
+ The goal of this experiment is to see if K562 cells can be stimulated by THP-1 Cells through Paracrine signalling.
+ We will use vented Tissue Culture Treated T12.5 Flasks for this experiment.
+ Four days before the experiment(preferably on a Friday) prepare 40ml of THP-1 cells 430cells/&mu;l.
    + Spin down the cells and use fresh complete growth media.
    + Resuspend the cells then 40&mu;m filter the cells.
    + Count the cells and resuspend at 430cells/&mu;l.
+ Load the cells into the ChronoSeq device and leave with Vent On with regular blowout to clear condensation every 2 hours for a total of three days.
    + Make sure there is a Magnetic Stirring disk in the Tube.
+ Add 0.4&mu;l of PMA 10mM stock.Press down till second stop. The magnetic stirring disk should ensure good mixing.
+ Make sure you have enough 5%CO<sub>2</sub> in the Gas Cylinder to last the entire duration of the experiment.
+ Take a 20&mu;l sample of cells every day to check for clumping.
+ One the third day we will spin down the cells and try to break clumps through trypsinization if necessary. Then:
    + We will resuspend the cells after washing twice with PMA-Free media. 
    + Resuspend at 430cells/ul and then leave for 1 day as a resting period.

### On Day of Experiment 
+ Make 6 K562 Flasks with DMEM 1%PEN-STREP with 0.9ml of K562 Cells in each:
    + We will label these flasks 3hr, 2hr, 1hr, 0hr, No Stimulation and TNF&alpha;
+ Make a 7th flask with 5ml of DMEM 1%PEN-STREP for later use for control experiments.
+ Make enough Lysis reagent for 7 samples and keep on ice.
+ Thaw LPS and TNF&alpha; Vials and then keep on ice.
    
### Cell Lysis
+ We will follow a protocol [similar to SingleShot Cell Lysis Kit](https://www.bio-rad.com/sites/default/files/webroot/web/pdf/lsr/literature/10000112808.pdf) but we will not change our media to DPBS.
+ Prepare 6 PCR Tubes for Cell Lysis as follows and keep these Tubes on ice while your prepare. For each tube:
    + Label the Tubes 3H,2H,1H,0H,NS,TNF.
    + 48&mu;l Single Shot Cell Lysis Buffer
    + 1&mu;l Proteinase K Solution
    + 1&mu;l DNase Solution
+ For each tube we will:
    + Add 5&mu;l of Cell Suspension. Mix Well. This is equivalent to 1075 K562 Cells.
    + Keep the tube back on ice till you have collected all 6 samples.
   
+ We need to take both 0hr and 1hr samples for K562 as well. This will help calculate the impact of residual THP-1 cells in the supernatant as well. We will only take 0.1ml Samples of THP-1 cells to conserve the amount left for our experiment.

|Time|THP-1 Sample|K562 5&mu;l Sample for qPCR|
|-|-|-|
|0hr|Add LPS take 0.1 ml sample spin down at 1500g for 1 minute and then add 50ul Supernatant to K562 0hr Epie|Immediately also take sample for K562|
|30min|Add 0.9ml of DMEM 1%PEN-STREP media to No Stimulation and TNF&alpha; flasks. Immediately add 1.8&mu;l of TNF&alpha; stock to TNF&alpha; flask| Immediately take 0hr sample for both K562 Tubes|
|1hr|take 0.1 ml sample spin down at 1500g for 1 minute and then add 50ul Supernatant to K562 1hr flask|Immediately Take sample form 1hr sample and then Take 5ul sample from K562 0hr flask for qPCR|
|1hr30mins|Do nothing|Take 5ul samples from No Stimulation and TNF&alpha K562 Flask for qPCR|
|2hr|take 0.1 ml sample spin down at 1500g for 1 minute and then add 50ul Supernatant to K562 2hr flask|Immediately take sample from 2hr tube then, Take 5ul sample from K562 1hr flask for qPCR|
|3hr|take 0.1 ml sample spin down at 1500g for 1 minute and then add 50ul Supernatant to K562 3hr flask|Immediately take sample from 3hr tube then, Take 5ul sample from K562 2hr flask for qPCR|
|4hr|Do nothing|Take 5ul sample from K562 3hr flask for qPCR|

+ Spin down all the Tubes.
    + Using a PCR Machine run the following program:
        + Lid Temperature 105&deg;C
        + Volume 55&mu;l
        + Incubate for 10 min at 21&deg;C
        + Followed by 5 min at 37°C,
        + 5 min at 75°C.
        + 4&deg;C for &infin;
    + The cell lysate can be stored on ice for up to 4 hr, at -20°C for up to 2 months, or at -80°C for up to 12 months.
    + After PCR Immediately store the Tubes at -80&deg;C

### Single Step RT-qPCR
+ Adapted [from existing protocol here](https://www.bio-rad.com/webroot/web/pdf/lsr/literature/10032048.pdf).
+ The iTaq kit comes with ROX as a background reference dye so remember to set that in your qPCR machine. ROX is not necessary for Bio-Rad CFX Opus 96. 
+ For each conditon you will have to setup 4 qPCR reactions. Two replicates for each gene NFKBIA and ACTB. For each 20&mu;l reaction PCR Tube:
    + iTaq™ Universal SYBR® Green reaction mix 2X: 10&mu;l
    + 1&mu;l PrimePCR Assay primers (20X).
    + iScript reverse transcriptase 0.25&mu;l
    + 4&mu;l Cell Lysate.
    + 4.75&mu;l Nuclease Free Water
+ Setup the qPCR reaction as follows: 
    + 10min at 50&deg;C for Reverse Transcription.
    + 1min at 95&deg;C for DNA denaturation and Polymerse activation.
    + 40 Cycles of:
        + 95&deg;C 10seconds Denaturation
        + 60&deg;C 30seconds Annealing and Plate Read
+ After the plate setup. The plate was sealed with a Optically clear adhesive film.
    + Make sure the lips are completely sealed to avoid leaks. Then vortex the plate.
+ Plate was then spun down and loaded into the qPCR machine.
+ Run the Program.
